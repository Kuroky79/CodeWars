```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pandas as pd
```

```python
# Задаем данные
training_data = pd.read_excel('data.xlsx')
print('data.xlsx')
```

```python
# Создание словаря для преобразования текстовых меток в числовые значения
label_to_index = {label: i for i, label in enumerate(set([data[1] for data in training_data]))}
index_to_label = {i: label for label, i in label_to_index.items()}

```

```python
# Создание обучающих данных и меток
train_X = np.array([data[0] for data in training_data])
train_y = np.array([label_to_index[data[1]] for data in training_data])
```

```python
# Создание токенизатора и преобразование текста в последовательности чисел
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_X)
train_sequences = tokenizer.texts_to_sequences(train_X)

```

```python
# Получение общего числа слов в словаре
vocab_size = len(tokenizer.word_index) + 1

```

```python
# Приведение всех последовательностей к одной и той же длине
max_sequence_length = max([len(seq) for seq in train_sequences])
train_X = pad_sequences(train_sequences, maxlen=max_sequence_length)

```

```python
# Создание модели
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 64, input_length=max_sequence_length),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(len(label_to_index), activation='softmax')
])
```

```python
# Компиляция модели
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

```

```python
# Обучение модели
model.fit(train_X, train_y, epochs=1000)
```

```python
# Функция для предсказания ОКПД по наименованию товара
def predict_okpd(product_name):
    product_sequence = tokenizer.texts_to_sequences([product_name])
    product_sequence = pad_sequences(product_sequence, maxlen=max_sequence_length)
    prediction = model.predict(np.array(product_sequence))
    predicted_label = np.argmax(prediction[0])
    return index_to_label[predicted_label]
```

```python
# Пример использования модели для предсказания ОКПД
product_name = "Услуги по аренде грузовых транспортных средств с водителем"
predicted_okpd = predict_okpd(product_name)
print(f"Наименование товара: {product_name}")
print(f"Предсказанный ОКПД: {predicted_okpd}")
```
