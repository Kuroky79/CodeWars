```python
data = pd.read_excel('traindataa.xlsx')
df = pd.read_excel('traindataa.xlsx')
X_train = data['Наименование']
y_train = data['Артикул'].values
traindata = df[['Наименование', 'Артикул']].values.tolist()
```

```python
df = pd.read_excel('testdataa.xlsx')
X_test = data['Наименование']
y_test = data['Артикул'].values
testdata = df[['Наименование', 'Артикул']].values.tolist()
```

```python
# Tokenize text data
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)
```

```python
from keras.utils import pad_sequences
# Pad sequences
maxlen = 100
X_train_pad = pad_sequences(X_train_seq, padding='post', maxlen=maxlen)
X_test_pad = pad_sequences(X_test_seq, padding='post', maxlen=maxlen)
```

```python
from sklearn.preprocessing import LabelEncoder
y_train = y_train.astype(str)
y_test = y_test.astype(str)
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train_bin = lb.fit_transform(y_train)
y_test_bin = lb.transform(y_test)
```

```python
# Define model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=10000, output_dim=128, input_length=maxlen),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2151, activation='softmax')

])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


```

```python
# Train model
model.fit(X_train_pad, y_train_bin, epochs=1000,validation_data=(X_test_pad, y_test_bin))
```

```python
# Тестирование модели
test_data = [
    'Кабель для монитора сетевой 1,8м (КБ-1.8)',
    'Exegate ES280987RUS Кабель монитор - компьютер (UPS-устройство) ExeGate Special EC-1,8S (IEC 320 С13-С14) 3*0.5mm2, CCA, черный, 1.8м.',
    'Filum Aura Спрей для очистки мониторов и оптики, 250 мл (CLN-S250ICD)'
]

# Tokenize test data
test_data_seq = tokenizer.texts_to_sequences(test_data)

# Pad sequences
test_data_pad = pad_sequences(test_data_seq, padding='post', maxlen=maxlen)

# Make predictions
predictions = model.predict(test_data_pad)

# Print results
for i, prediction in enumerate(predictions):
    max_prob = max(prediction)
    if max_prob >= 0.5:
        index = np.argmax(prediction)
        print(f'Артикул для {test_data[i]}: {traindata[index][1]}')
    else:
        print(f'Для {test_data[i]} артикул не найден')
```
